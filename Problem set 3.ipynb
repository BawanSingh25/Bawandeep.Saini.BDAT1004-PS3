# Question 6 
## Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend. Don't forget to label your axes!

```python
# importing pandas

import pandas as pd

# reading data from csv file

data = pd.read_csv("us-marriages-divorces-1867-2014.csv") 
data.head()

# importing seborn 

import seaborn as sns

#to calculate marriages per capita

data['marriages_pc'] = data['Marriages']/data['Population'] 

 #to calculate divorces per capita


data["divorces_pc"]  = data['Divorces']/data['Population']    

#using required attributes only

data2 = data[['Year','marriages_pc','divorces_pc']]  

#between 1867 & 2014
    
data2 = data2.loc[(data2['Year']>=1867)&(data2['Year']<=2014)]   
ax = sns.lineplot('Year', 'value',color = 'red', hue='variable',
             data=pd.melt(data2, 'Year'))
ax.set_title('number of marriages and divorces per capita in the U.S​ \n 1867 - 2014')
```
